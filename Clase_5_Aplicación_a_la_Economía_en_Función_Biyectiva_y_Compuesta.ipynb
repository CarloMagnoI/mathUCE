#### üìö***Definici√≥n y Representaci√≥n de Funciones***

Una **funci√≥n** es una regla de asociaci√≥n que asigna a cada elemento de un conjunto de partida $( A )$ un elemento √∫nico en un conjunto de llegada $( B )$. La notaci√≥n matem√°tica est√°ndar para una funci√≥n es $( f: A \to B )$. Visualmente, podemos imaginar esto como un mapa que toma cada elemento $( x )$ en $( A )$ y lo asocia con un elemento $( f(x) )$ en $( B )$.

Imaginemos que $( A )$ es un conjunto de n√∫meros y $( B )$ es un conjunto de colores. La funci√≥n $( f )$ podr√≠a representar una regla que asocia cada n√∫mero con un color espec√≠fico.





#### üìö***Imagen y Preimagen de una Funci√≥n***

Cuando hablamos de funciones, es crucial entender los conceptos de **imagen** y **preimagen**:

- **Preimagen**: La preimagen de un conjunto en $( B )$ es el conjunto de todos los elementos en $( A )$ que se mapean a ese conjunto espec√≠fico en $( B )$ bajo la funci√≥n $( f )$. La preimagen de un elemento $( y )$ en $( B )$ se denota como $( f^{-1}(y) )$. $(Dominio)$.
- **Imagen**: La imagen de un conjunto $( A )$ bajo una funci√≥n $( f )$ es el conjunto de todos los valores que $( f(x) )$ puede tomar cuando $( x )$ var√≠a en $( A )$. Matem√°ticamente, se denota como $( f(A) )$. $(Rango)$ y/o $(Recorrido)$.




## üë©‚Äçüî¨ ***Funciones de Utilidad en Econom√≠a***

En el √°mbito econ√≥mico, una **funci√≥n de utilidad** es una herramienta que se utiliza para representar el nivel de satisfacci√≥n o bienestar de un individuo en relaci√≥n con el consumo de bienes. Esta funci√≥n toma como entrada la cantidad de bienes consumidos y produce como salida un valor que representa la utilidad o satisfacci√≥n obtenida.



- ##### üí≠ ***Funci√≥n de Utilidad con un Solo Bien***

Consideremos un ejemplo simple en el que el √∫nico bien disponible es agua. La funci√≥n de utilidad, en este caso, puede representarse como $( u(x) )$, donde $( x )$ es la cantidad de agua consumida y $( u(x) )$ es la utilidad obtenida.

$[ \text{Agua} \, (x) \rightarrow \text{Utilidad} \, (u(x)) $]

Gr√°ficamente, podemos representar esta relaci√≥n con una curva que muestra c√≥mo la utilidad aumenta a medida que se consume m√°s agua. Sin embargo, la utilidad marginalmente decrece, lo que significa que cada unidad adicional de agua proporciona una utilidad menor que la anterior. Por ejemplo, el primer vaso de agua puede proporcionar mucha satisfacci√≥n en un d√≠a caluroso, pero el tercer o cuarto vaso proporciona una satisfacci√≥n significativamente menor.



- ##### üí≠ ***Funciones de Utilidad con Dos Bienes***

La situaci√≥n se vuelve m√°s interesante cuando consideramos dos bienes, como agua y pan. Ahora, la funci√≥n de utilidad depende de ambas cantidades consumidas:

$[ u(x, y) ]$

Aqu√≠, $( x )$ es la cantidad de agua y $( y )$ es la cantidad de pan. La utilidad se representa en un gr√°fico tridimensional, donde los ejes $( x )$ e $( y )$ corresponden a las cantidades de los dos bienes, y el eje vertical representa la utilidad. Esta representaci√≥n gr√°fica muestra una **superficie de utilidad**.

Imaginemos una superficie en forma de concha ac√∫stica, donde cada punto en la superficie representa un nivel de utilidad espec√≠fico asociado con una combinaci√≥n particular de $( x )$ e $( y )$. Si solo miramos desde el lado de $( x )$ o $( y )$, veremos curvas similares a las que ten√≠amos en el caso unidimensional.

Ejemplos de funciones de utilidad que se ajustan a esta descripci√≥n incluyen:

- $( \log(x) )$
- $( \sqrt{x} )$

Ambas funciones tienen la caracter√≠stica de ser crecientes pero con una pendiente cada vez menor, reflejando el principio de ***utilidad marginal decreciente.***

- ## üî¢ ***Notaci√≥n Matem√°tica de la Funci√≥n de Utilidad***

1. **Definici√≥n de la Funci√≥n de Utilidad**:
   $[
   U: \mathbb{R}^2_+ \rightarrow \mathbb{R}^+
   ]$
   Esta notaci√≥n indica que la funci√≥n de utilidad \( U \) toma como entrada un par de n√∫meros reales no negativos $(( \mathbb{R}^2_+ ))$ y produce un n√∫mero real no negativo $(( \mathbb{R}^+ ))$.

2. **Expresi√≥n de la Funci√≥n de Utilidad**:
   $[
   (x, y) \rightarrow u(x, y)
   ]$
   Aqu√≠, $( (x, y) )$ son las cantidades de dos bienes, y $( u(x, y) )$ es el nivel de utilidad asociado con esas cantidades.

3. **Conjuntos de Nivel**:
   $[
   \{ (x, y) \in \mathbb{R}^2_+ : u(x, y) = 10 \}
   ]$
   Esto representa el conjunto de todos los pares $((x, y))$ en $(\mathbb{R}^2_+)$ que producen una utilidad de 10. En el gr√°fico, estos conjuntos de nivel est√°n representados por curvas de nivel $(isocuantas)$ en rojo

- ***Gr√°fica de la Funci√≥n en Python***

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la funci√≥n de utilidad
def u(x, y):
    return np.log(x) + np.sqrt(y)

# Rango de valores para x e y
x = np.linspace(0.1, 5, 400)  # evitamos x=0 para evitar log(0)
y = np.linspace(0, 10, 400)
X, Y = np.meshgrid(x, y)
Z = u(X, Y)

# Creamos la figura
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Gr√°fico de la superficie de la funci√≥n de utilidad
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.8)

# Contornos en el plano
contours = ax.contour(X, Y, Z, levels=[1, 2, 3, 4, 5, 10], cmap='autumn', linestyles="solid")

# Etiquetas y t√≠tulo
ax.set_xlabel('Bien X')
ax.set_ylabel('Bien Y')
ax.set_zlabel('Utilidad U(x, y)')
ax.set_title('Superficie de la Funci√≥n de Utilidad y Conjuntos de Nivel')

# Mostrar el gr√°fico
plt.show()
```

 ***Continuaci√≥n de la Funci√≥n de Utilidad: Conjuntos de Nivel y Preimagen***

**Notaci√≥n Matem√°tica**

1. **Conjuntos de Nivel**:

   $[
   f^{-1}(\{10\}) = \{ (x, y) \in \mathbb{R}^2_+ : u(x, y) \in \{10\} \}
   ]$

   Este conjunto representa todos los pares $((x, y))$ que producen una utilidad de 10. Esto se expresa de la siguiente manera:

   $[
   \{ (x, y) : \sqrt{xy} = 10 \}
   ]$

   Simplificando, obtenemos:

   $[
   \{ (x, y) : xy = 100 \}
   ]$

   O, de manera equivalente:

   $[
   \{ (x, y) : y = \frac{100}{x} \}
   ]$

2. **Generalizaci√≥n para un Nivel de Utilidad $( u_0 )$**:

   $[
   f^{-1}(\{u_0\}) = \{ (x, y) \in \mathbb{R}^2_+ : y = \frac{u_0^2}{x} \}
   ]$

   Esto indica que, para un nivel de utilidad $( u_0 )$, la relaci√≥n entre $( x )$ e $( y )$ se mantiene conforme a la ecuaci√≥n mencionada.
   


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definimos la funci√≥n de utilidad
def u(x, y):
    return np.sqrt(x * y)

# Rango de valores para x e y
x = np.linspace(0.1, 10, 400)  # evitamos x=0 para evitar problemas de log(0)
y = np.linspace(0.1, 10, 400)
X, Y = np.meshgrid(x, y)
Z = u(X, Y)

# Valores de utilidad para los contornos
u_levels = [5, 10, 15, 20, 25]

# Creamos la figura
fig = plt.figure(figsize=(14, 9))
ax = fig.add_subplot(111, projection='3d')

# Gr√°fico de la superficie de la funci√≥n de utilidad
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.7)

# Dibujamos los conjuntos de nivel
for u0 in u_levels:
    x_contour = np.linspace(0.1, 10, 400)
    y_contour = (u0**2) / x_contour
    z_contour = u(x_contour, y_contour)
    ax.plot(x_contour, y_contour, z_contour, label=f'u0 = {u0}', linewidth=2)

# Etiquetas y t√≠tulo
ax.set_xlabel('Bien X')
ax.set_ylabel('Bien Y')
ax.set_zlabel('Utilidad U(x, y)')
ax.set_title('Superficie de la Funci√≥n de Utilidad y Conjuntos de Nivel')

# Mostrar leyenda y gr√°fico
ax.legend()
plt.show()
```

***El gr√°fico en pocas plabras representa, la Utilidad Marginal Decreciente (beneficio de producir, consumir, etc, una unidad mas), y las Curvas de Indiferncia, (Conjuntos de bienes que son preferibles a otros segun las diferentes condiciones).***

- ***Curvas de indiferencia de la Funci√≥n de Utilidad***

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la funci√≥n de utilidad y su inversa para los conjuntos de nivel
def u(x, y):
    return np.sqrt(x * y)

def preimage(u0, x):
    return u0**2 / x

# Valores de utilidad para los contornos
u_levels = [5, 10, 15, 20, 25]

# Rango de valores para x
x = np.linspace(0.1, 10, 400)  # evitamos x=0 para evitar divisi√≥n por cero

# Creamos la figura
plt.figure(figsize=(10, 6))

# Dibujamos los conjuntos de nivel
for u0 in u_levels:
    y = preimage(u0, x)
    plt.plot(x, y, label=f'u0 = {u0}')

# Etiquetas y t√≠tulo
plt.xlabel('Bien X')
plt.ylabel('Bien Y')
plt.title('Conjuntos de Nivel de la Funci√≥n de Utilidad')
plt.legend()

# Mostrar el gr√°fico
plt.grid(True)
plt.show()
```

## üíª ***Funci√≥n Sobreyectiva, Biyectiva, Inyectiva, Inversa y Compuesta***



#### ‚úîÔ∏è  1. ***Definici√≥n de una Funci√≥n***
Una funci√≥n $( f )$ que va del conjunto $( A )$ al conjunto $( B )$ se denota como:
$[ f: A \to B ]$
y asigna a cada elemento $( x \in A )$ un elemento $( f(x) \in B )$.



#### ‚úîÔ∏è ***2. Im√°genes de una Funci√≥n***
- La imagen de $( A )$ bajo $( f )$, denotada como $( f(A) )$, es el conjunto de todos los elementos de $( B )$ que son im√°genes de los elementos de $( A ):
  [ f(A) = \{ f(x) \in B : x \in A \} ]$



#### ‚úîÔ∏è***3. Sobreyectividad***
- Una funci√≥n $( f )$ es **sobreyectiva** si su imagen cubre todo el conjunto $( B ):
  [ f(A) = B ]$

  Esto significa que para cada $( y \in B )$, existe al menos un $( x \in A )$ tal que:

  $[ f(x) = y ]$


  
#### ‚úîÔ∏è ***4. Inyectividad***
- Una funci√≥n $( f )$ es **inyectiva** si asigna elementos distintos de $( A )$ a elementos distintos de $( B )$. Formalmente, $( f )$ es inyectiva si:

  $[ \forall x, x' \in A, \, x \neq x' \implies f(x) \neq f(x') ]$



#### ‚úîÔ∏è ***5. Biyectividad***
- Una funci√≥n $( f )$ es **biyectiva** si es a la vez inyectiva y sobreyectiva. Esto implica que $( f )$ asigna una correspondencia uno a uno entre los elementos de $( A )$ y los elementos de $( B )$. En t√©rminos formales:
  $[ f \text{ es biyectiva} \iff f \text{ es inyectiva} \, \land \, f \text{ es sobreyectiva} ]$



#### ‚úîÔ∏è ***6. Funci√≥n Inversa***
- Si $( f )$ es biyectiva, entonces tiene una funci√≥n inversa $( f^{-1} )$ que va de $( B )$ a $( A )$.

  La funci√≥n inversa $( g )$ cumple con:
  $[ g: B \to A ]$
  $[ g(y) = x \quad \text{tal que} \quad f(x) = y ]$

  La funci√≥n inversa $( g )$ se denota como $( f^{-1} )$.
  
   Las propiedades que deben cumplirse son:
  $[ \forall x \in A, \, g(f(x)) = x ]$
  $[ \forall y \in B, \, f(g(y)) = y ]$



```python
import matplotlib.pyplot as plt
import numpy as np

# Crear una figura con tres subgr√°ficos
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Datos para las gr√°ficas
x = np.linspace(-10, 10, 400)

# Inyectiva: f(x) = 2x + 1
f_inyectiva = 2 * x + 1
axs[0].plot(x, f_inyectiva, label='f(x) = 2x + 1')
axs[0].set_title('Funci√≥n Inyectiva')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].axhline(0, color='black', linewidth=0.5)
axs[0].axvline(0, color='black', linewidth=0.5)
axs[0].legend()
axs[0].grid(True)

# Sobreyectiva: f(x) = tanh(x)
f_sobreyectiva = np.tanh(x)
axs[1].plot(x, f_sobreyectiva, label='f(x) = tanh(x)', color='orange')
axs[1].set_title('Funci√≥n Sobreyectiva')
axs[1].set_xlabel('x')
axs[1].set_ylabel('f(x)')
axs[1].axhline(0, color='black', linewidth=0.5)
axs[1].axvline(0, color='black', linewidth=0.5)
axs[1].legend()
axs[1].grid(True)

# Biyectiva: f(x) = x^3
f_biyectiva = x**3
axs[2].plot(x, f_biyectiva, label='f(x) = x^3', color='green')
axs[2].set_title('Funci√≥n Biyectiva')
axs[2].set_xlabel('x')
axs[2].set_ylabel('f(x)')
axs[2].axhline(0, color='black', linewidth=0.5)
axs[2].axvline(0, color='black', linewidth=0.5)
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

## ¬©Ô∏è ***Explicaci√≥n sobre Funciones Compuestas***



1. üîë **Definiciones Iniciales**:
  - *Se definen tres conjuntos:* $( E )$, $( F )$ y $( G )$.
  - Se definen dos funciones: \( f \) y \( g \).
    - $( f: E \rightarrow F \), donde \( f(x) = f(x) )$.

    - $( g: F \rightarrow G ), donde ( g(y) = g(y) )$.



2. üîë **Funci√≥n Compuesta**:
    - La funci√≥n compuesta
  $( g \circ f )$ se define como:
$[
(g \circ f)(x) = g(f(x))
]$
Esta composici√≥n toma un elemento $( x )$ de $( E )$, aplica la funci√≥n $( f )$ para obtener $( f(x) )$ en $( F )$, y luego aplica la funci√≥n $( g )$ a $( f(x) )$ para obtener $( g(f(x)) )$ en $( G )$.

3. üîë **Ejemplo de Funci√≥n Compuesta**:
  - Se define una funci√≥n $( h(x) \):
      \[
      h(x) = \sqrt{x^2 + 1}
      ]$


4. üîë **Definici√≥n de Funciones**:

    - Se define la funci√≥n $( f )$ de los reales a un intervalo:
      $[
      f: \mathbb{R} \rightarrow [1, +\infty[
      ]$
      $[
      f(x) = x^2 + 1
      ]$
    - Se define la funci√≥n $( g )$ del intervalo $( [1, +\infty[ )$ a los reales positivos:
      $[
      g: [1, +\infty[ \rightarrow \mathbb{R}^+
      ]$
      $[
      g(y) = \sqrt{y}
      ]$




5. üîë **Composici√≥n de $( g )$ y $( f )$**:

    - La composici√≥n $( g \circ f )$ se define como:
      $[
      g \circ f: \mathbb{R} \rightarrow \mathbb{R}^+
      ]$
      $[
      (g \circ f)(x) = g(f(x)) = g(x^2 + 1) = \sqrt{x^2 + 1}
      \]rt{x^2 + 1}



6. üîë **Conclusi√≥n**:
    - Se demuestra que $( h(x) )$ es igual a la composici√≥n $( g \circ f )$ para todos los $( x )$ en los reales:
      $[
      h(x) = g \circ f(x) \quad \forall x \in \mathbb{R} \Leftrightarrow h = g \circ f
      ]$

```python
import matplotlib.pyplot as plt
import numpy as np

# Definici√≥n de las funciones
def f(x):
    return x**2 + 1

def g(y):
    return np.sqrt(y)

def h(x):
    return g(f(x))

# Valores de x para graficar
x = np.linspace(-10, 10, 400)

# Graficar las funciones
plt.figure(figsize=(10, 6))

plt.plot(x, f(x), label='$f(x) = x^2 + 1$', color='blue')
plt.plot(x, g(f(x)), label='$g(f(x)) = \sqrt{x^2 + 1}$', color='red')
plt.plot(x, h(x), label='$h(x) = g(f(x)) = \sqrt{x^2 + 1}$', linestyle='dashed', color='green')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.title('Visualizaci√≥n de la Funci√≥n Compuesta $h(x) = g(f(x))$')

plt.show()
```

##  üåü ***Conclusi√≥n General sobre Funciones y su Aplicaci√≥n***

Las funciones inyectivas, sobreyectivas y biyectivas üéØ juegan un papel crucial en la teor√≠a de funciones, proporcionando una base s√≥lida para comprender c√≥mo los elementos de diferentes conjuntos est√°n relacionados entre s√≠. Estas propiedades son fundamentales para asegurar que cada elemento de un conjunto tiene una correspondencia clara y √∫nica en otro conjunto, lo cual es esencial en muchas aplicaciones matem√°ticas y cient√≠ficas üß†üìê.

En el √°mbito de la econom√≠a üìä, estas funciones se aplican para modelar relaciones entre variables econ√≥micas, como la utilidad y el consumo, permitiendo un an√°lisis preciso de c√≥mo los cambios en una variable afectan a otra. Las funciones compuestas üîÑ, que combinan varias funciones en una, son especialmente √∫tiles para describir procesos econ√≥micos complejos donde m√∫ltiples etapas de transformaci√≥n o cambio est√°n involucradas.

Mediante la composici√≥n de funciones, podemos simplificar y resolver problemas que implican varias transformaciones sucesivas, ofreciendo una herramienta poderosa para la modelizaci√≥n y el an√°lisis econ√≥mico üí°üíº. La visualizaci√≥n de estas funciones y sus composiciones no solo facilita la comprensi√≥n de sus propiedades te√≥ricas, sino que tambi√©n proporciona una representaci√≥n clara y accesible de sus aplicaciones pr√°cticas en econom√≠a y otros campos üìà.
