* *Lógica Proposicional:*

  ***Definición de Proposición:***
Una **proposición** es una afirmación que puede ser verdadera o falsa, pero no ambas.

  ***Operadores Lógicos:***
Los operadores lógicos principales son:

  - **Negación** (¬), niega una proposición
  - **Conjunción** (∧), se conoce como (y)
  - **Disyunción** (∨), se conoce como (o)
  - **Implicación** (→),  es una afirmación que conlleva otra.
  - **Doble implicación** (↔), bicondicional

  ***Cuantificadores***
  - **Universalidad** (∀), "Para todo"
  - **Existencia** (∃), "Existe"

  ***Ejemplos*** :

   *∃x ∀y:* "Existe un x, para todo y"

  ∀y ∃x: "Para todo y, existe un x"

    - Importante: El orden de los cuantificadores afecta el significado lógico.

  ***Analogía***:

 "Todos tienen una madre, ninguna como la mía"

    - ∃x ∀y: Existe una madre para todas las personas (incorrecto)
 - ∀y ∃x: Para toda persona existe ure (correcto)

***Negación de cuantificadores***

  - Negación de ∃x P(x): ∀x ¬P(x)
- Negación de ∀x P(x): ∃x ¬P(x)

*   *Teoría de Conjuntos:*

  ***Definición de Conjunto:***
  Un **conjunto** es una colección de elementos distintos. Los elementos pueden ser números, letras, u otros objetos.

  ***Notación:***

  - Un conjunto se denota con llaves: {a, b, c}
  - Conjunto vacío: {}

  ***Operaciones con Conjuntos:***

  - **Unión** (A ∪ B)
  - **Intersección** (A ∩ B)
  - **Diferencia** (A - B)
  - **Complemento** (A^c)

  ***Conjunto de los números:***
  - **ℕ, (números naturales):** {0,1,2,3,...,+∞}
  - **ℤ,(números enteros):** {-∞,...,-3,-2,-1,0,1,2,3,...,+∞}
  - **ℚ,(números racionales):** {n/;n ∈ Z, m ∈ N}
  - **ℝ,(números reales):** {toda la recta númerica}
  
    *Por ende*, **N ∈ Z ∈ Q ∈ R**
